```python
import subprocess
import json

# Définir votre commande curl
curl_command = '''
curl -X POST "http://localhost:8000/extract" \
  -H "Content-Type: application/json" \
  -d '{
    "urls": "https://www.m-lai.com"
  }'
'''

# Exécuter la commande curl
result = subprocess.run(curl_command, shell=True, capture_output=True, text=True)

# Afficher la sortie
print("Statut:", result.returncode)
print("Sortie standard:")
print(result.stdout)

# En cas d'erreur
if result.returncode != 0:
    print("Erreur:")
    print(result.stderr)

# Si la sortie est au format JSON, vous pouvez la parser
try:
    json_response = json.loads(result.stdout)
    print("\nRéponse JSON formatée:")
    print(json.dumps(json_response, indent=2))
except json.JSONDecodeError:
    print("\nLa sortie n'est pas au format JSON valide")
```
